```python
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train=x_train.reshape(60000,28,28,1)
```

```python
x_train=x_train/255
```

```python
#cnn
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
```

```python
# building the CNN model
model = Sequential()

model.add(Conv2D(64, kernel_size=(3,3), input_shape=(28,28,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='softmax'))
#for binary sigmoid #for regression linear  #for multiple classes softmax
```

```python
model.summary()
```

```python
Model: 'sequential'
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossntropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10,validation_split=0.1)
```
