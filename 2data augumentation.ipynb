```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/Users/vishwamalani/Desktop/python/DA/CNN/train/cat.jpeg',target_size=(200,200))
```

```python

from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40, #angle
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest') #it can be nearest , reflect,constant
```

```python
img = image.img_to_array(img) 
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1,save_to_dir='/Users/vishwamalani/Desktop/python/DA/CNN/preview', save_prefix='cat', save_format='jpeg'):
    i += 1
    if i == 10:
        break 
```
